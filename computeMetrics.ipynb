```python
# Cell 6: Arabic Text Processing Class
class ArabicTextProcessor:
    """Class for processing Arabic text"""
    
    def __init__(self):
        self.stopwords = set(stopwords_list())
        self.stemmer = ISRIStemmer()
        
    def clean_text(self, text: str) -> str:
        """Clean Arabic text"""
        if pd.isna(text):
            return ""
        
        text = str(text)
        # Normalize Arabic characters
        text = araby.normalize_hamza(text)
        text = araby.normalize_alef(text)
        
        # Remove diacritics
        text = araby.strip_tashkeel(text)
        
        # Remove extra spaces and newlines
        text = re.sub(r'\s+', ' ', text).strip()
        
        return text
    
    def tokenize(self, text: str) -> List[str]:
        """Tokenize Arabic text"""
        text = self.clean_text(text)
        tokens = araby.tokenize(text)
        # Remove stopwords
        tokens = [t for t in tokens if t not in self.stopwords and len(t) > 1]
        return tokens
    
    def get_root(self, word: str) -> str:
        """Get morphological root of Arabic word"""
        word = self.clean_text(word)
        return self.stemmer.stem(word)
    
    def normalize_arabic_word(self, word: str) -> str:
        """Normalize Arabic word for comparison"""
        word = self.clean_text(word)
        # Remove all non-Arabic characters
        word = re.sub(r'[^\u0600-\u06FF]', '', word)
        return word

# Initialize text processor
processor = ArabicTextProcessor()
```

```python
# Cell 9: Evaluation Metrics Implementation
from typing import List, Dict
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

class ArabicLMEvaluator:
    """Evaluator for Arabic language models"""
    
    def __init__(self, processor: ArabicTextProcessor, arabic_wordnet=None):
        """
        Args:
            processor: ArabicTextProcessor instance
            arabic_wordnet: Your Arabic WordNet object (from wn.Wordnet('omw-arb:1.4'))
        """
        self.processor = processor
        self.arabic_wordnet = arabic_wordnet
        
    def accuracy_score(self, predicted: str, expected: str) -> float:
        """
        Accuracy: Exact match with expected words
        Expected can contain multiple possibilities separated by spaces
        Returns: 1.0 if match, 0.0 otherwise
        """
        if pd.isna(predicted) or pd.isna(expected):
            return 0.0
        
        predicted_norm = self.processor.normalize_arabic_word(predicted)
        expected_words = [self.processor.normalize_arabic_word(w) for w in str(expected).split()]
        
        if predicted_norm in expected_words:
            return 1.0
        return 0.0
        
    def _get_wordnet_synonyms(self, word: str) -> List[str]:
        """Get Arabic synonyms from Arabic WordNet using wn 0.14.0."""
        word = self.processor.normalize_arabic_word(word)
        synonyms = set()

        if self.arabic_wordnet is None:
            print("Warning: Arabic WordNet not available")
            return [word]

        try:
            synsets = self.arabic_wordnet.synsets(word)

            for syn in synsets:
                # In wn 0.14.0, syn.lemmas() returns strings directly
                for lemma_str in syn.lemmas():
                    lemma_norm = self.processor.normalize_arabic_word(lemma_str)
                    synonyms.add(lemma_norm)

        except Exception as e:
            print(f"Warning: WordNet lookup failed for '{word}': {e}")

        synonyms.add(word)
        return list(synonyms)
        
    def synonym_score(self, predicted: str, expected: str) -> float:
        """
        Synonym Match: Checks if predicted word or its synonyms match expected words
        using real Arabic WordNet.
        Normalizes and reduces both predicted and expected words and synonyms to roots.
        Returns: 1.0 if match, 0.0 otherwise
        """
        if pd.isna(predicted) or pd.isna(expected):
            return 0.0

        # Normalize and reduce predicted to root
        predicted_root = self.processor.get_root(predicted)

        # Split expected words, normalize, and reduce to roots
        expected_roots = [self.processor.get_root(w) for w in str(expected).split()]

        # Check direct root match
        if predicted_root in expected_roots:
            return 1.0

        # Get WordNet synonyms for predicted word
        synonyms = self._get_wordnet_synonyms(predicted)

        # Normalize and reduce synonyms to roots
        synonym_roots = [self.processor.get_root(s) for s in synonyms]

        # Check if any synonym root matches expected roots
        for root in synonym_roots:
            if root in expected_roots and root != "":
                return 1.0

        return 0.0
    
    def root_match_score(self, predicted: str, expected: str) -> float:
        """
        Root Match: Compares morphological roots
        Returns: 1.0 if roots match, 0.0 otherwise
        """
        if pd.isna(predicted) or pd.isna(expected):
            return 0.0
        
        predicted_root = self.processor.get_root(predicted)
        expected_words = str(expected).split()
        
        for exp_word in expected_words:
            exp_root = self.processor.get_root(exp_word)
            if predicted_root == exp_root and predicted_root != "":
                return 1.0
        
        return 0.0
    
    def cosine_similarity_score(self, predicted: str, expected: str, vector_func=None) -> float:
        """
        Cosine Similarity: Measures semantic similarity
        Args:
            predicted: Predicted word
            expected: Expected word(s)
            vector_func: Function to get word vectors (default: get_word_vector)
        Returns: Cosine similarity score (0.0 to 1.0)
        """
        if pd.isna(predicted) or pd.isna(expected):
            return 0.0
        
        # Use provided vector function or default
        vector_func = vector_func or get_word_vector
        
        # Get vectors for predicted word
        try:
            pred_vector = vector_func(predicted).reshape(1, -1)
        except:
            return 0.0
        
        # Calculate max similarity with any expected word
        expected_words = str(expected).split()
        max_similarity = 0.0
        
        for exp_word in expected_words:
            try:
                exp_vector = vector_func(exp_word).reshape(1, -1)
                
                # Ensure vectors are not all zeros
                if np.any(pred_vector) and np.any(exp_vector):
                    similarity = cosine_similarity(pred_vector, exp_vector)[0][0]
                    # Clip to [0, 1] range
                    similarity = max(0.0, min(1.0, similarity))
                    max_similarity = max(max_similarity, similarity)
            except:
                continue
        
        return max_similarity
    
    def evaluate_row(self, row: pd.Series, vector_func=None) -> Dict[str, float]:
        """Evaluate a single row"""
        predicted = row['predicted_token']
        expected = row['المتوقع']
        
        return {
            'accuracy': self.accuracy_score(predicted, expected),
            'synonym_match': self.synonym_score(predicted, expected),
            'root_match': self.root_match_score(predicted, expected),
            'cosine_similarity': self.cosine_similarity_score(predicted, expected, vector_func)
        }
    
    def evaluate_dataset(self, df: pd.DataFrame, vector_func=None, verbose: bool = True, max_warnings: int = 10) -> pd.DataFrame:
        """Evaluate entire dataset"""
        if verbose:
            print(f"Evaluating {len(df)} rows...")
        
        # Counter for warnings to avoid spam
        warning_count = 0
        
        # Apply evaluation to each row
        results = []
        for idx, row in df.iterrows():
            scores = self.evaluate_row(row, vector_func)
            scores['index'] = idx
            results.append(scores)
            
            # Track warnings
            if warning_count < max_warnings:
                # You can add warning tracking logic here if needed
                pass
        
        # Create results dataframe
        results_df = pd.DataFrame(results)
        
        # Add scores to original dataframe
        for metric in ['accuracy', 'synonym_match', 'root_match', 'cosine_similarity']:
            df[f'{metric}_score'] = results_df[metric]
        
        # Calculate overall scores
        if verbose:
            print("\nEvaluation Results:")
            for metric in ['accuracy', 'synonym_match', 'root_match', 'cosine_similarity']:
                avg_score = df[f'{metric}_score'].mean()
                print(f"{metric}: {avg_score:.4f}")
        
        return df

    def get_summary_statistics(self, df: pd.DataFrame) -> pd.DataFrame:
        """Get summary statistics of all evaluation metrics"""
        summary_data = []
        for metric in ['accuracy', 'synonym_match', 'root_match', 'cosine_similarity']:
            col_name = f'{metric}_score'
            if col_name in df.columns:
                metric_data = {
                    'metric': metric,
                    'mean': df[col_name].mean(),
                    'median': df[col_name].median(),
                    'std': df[col_name].std(),
                    'min': df[col_name].min(),
                    'max': df[col_name].max(),
                    'count': df[col_name].count(),
                    'non_zero': (df[col_name] > 0).sum(),
                    'perfect_matches': (df[col_name] == 1.0).sum()
                }
                summary_data.append(metric_data)
        
        return pd.DataFrame(summary_data)

# Initialize evaluator with your existing arabic_wordnet object
evaluator = ArabicLMEvaluator(processor, arabic_wordnet=arabic_wordnet)
print("Arabic WordNet evaluator initialized successfully!")
print("Note: Using wn 0.14.0 compatibility mode")
```

```python
# Cell 11: Calculate Overall Performance Metrics
def calculate_overall_metrics(df_evaluated: pd.DataFrame) -> pd.DataFrame:
    """Calculate overall performance metrics"""
    
    metrics = ['accuracy_score', 'synonym_match_score', 'root_match_score', 'cosine_similarity_score']
    
    results = {}
    for metric in metrics:
        if metric in df_evaluated.columns:
            mean_score = df_evaluated[metric].mean()
            std_score = df_evaluated[metric].std()
            results[metric] = {
                'Mean': mean_score,
                'Std': std_score,
                'Count': df_evaluated[metric].count(),
                'Non-zero': (df_evaluated[metric] > 0).sum()
            }
    
    # Create summary dataframe
    summary_df = pd.DataFrame(results).T
    
    # Add percentage columns
    summary_df['Success_Rate'] = summary_df['Non-zero'] / summary_df['Count'] * 100
    summary_df['Mean_Percentage'] = summary_df['Mean'] * 100
    
    # Format for better display
    summary_df['Mean'] = summary_df['Mean'].round(4)
    summary_df['Std'] = summary_df['Std'].round(4)
    summary_df['Success_Rate'] = summary_df['Success_Rate'].round(2)
    summary_df['Mean_Percentage'] = summary_df['Mean_Percentage'].round(2)
    
    return summary_df

# Calculate overall metrics
summary_df = calculate_overall_metrics(df_evaluated)

print("Overall Performance Summary:")
print("=" * 80)
display(summary_df)
```

```python
# Cell 12: Performance by Category Analysis
def analyze_by_category(df_evaluated: pd.DataFrame, category_col: str = 'category2'):
    """Analyze performance by category"""
    
    if category_col not in df_evaluated.columns:
        print(f"Category column '{category_col}' not found. Available columns: {list(df_evaluated.columns)}")
        
        # Try to find any category column
        category_cols = [col for col in df_evaluated.columns if 'category' in col.lower()]
        if category_cols:
            category_col = category_cols[0]
            print(f"Using column: {category_col}")
        else:
            print("No category column found. Skipping category analysis.")
            return None
    
    metrics = ['accuracy_score', 'synonym_match_score', 'root_match_score', 'cosine_similarity_score']
    
    # Group by category
    category_stats = df_evaluated.groupby(category_col)[metrics].agg(['mean', 'count']).round(4)
    
    # Flatten column names
    category_stats.columns = ['_'.join(col).strip() for col in category_stats.columns.values]
    
    # Calculate success rates
    for metric in metrics:
        category_stats[f'{metric}_success_rate'] = (
            df_evaluated.groupby(category_col)[metric].apply(lambda x: (x > 0).mean() * 100)
        ).round(2)
    
    print(f"Performance by {category_col}:")
    print("=" * 80)
    display(category_stats)
    
    return category_stats

# Analyze by category if available
category_stats = analyze_by_category(df_evaluated)
```

```python
# Analyze performance by category1 even when multiple categories per cell
def analyze_by_category_multi(df_evaluated: pd.DataFrame, category_col: str = 'category1'):
    """
    Analyze performance by category.
    Handles cells that may contain multiple categories separated by space or comma.
    """
    
    if category_col not in df_evaluated.columns:
        print(f"Category column '{category_col}' not found. Available columns: {list(df_evaluated.columns)}")
        return None
    
    metrics = ['accuracy_score', 'synonym_match_score', 'root_match_score', 'cosine_similarity_score']
    
    # Split categories by space (or comma if needed)
    df_expanded = df_evaluated.copy()
    df_expanded[category_col] = df_expanded[category_col].astype(str).str.replace(',', ' ')
    df_expanded = df_expanded.assign(
        category_split = df_expanded[category_col].str.split()
    ).explode('category_split')
    
    # Group by individual category
    category1_stats = df_expanded.groupby('category_split')[metrics].agg(['mean', 'count']).round(4)
    
    # Flatten column names
    category1_stats.columns = ['_'.join(col).strip() for col in category1_stats.columns.values]
    
    # Calculate success rates
    for metric in metrics:
        category1_stats[f'{metric}_success_rate'] = (
            df_expanded.groupby('category_split')[metric].apply(lambda x: (x > 0).mean() * 100)
        ).round(2)
    
    print(f"Performance by {category_col}:")
    print("=" * 80)
    display(category1_stats)
    
    return category1_stats

# Run
category1_stats = analyze_by_category_multi(df_evaluated)

```

```python
# Cell 15: Error Analysis (wrong predictions only)
def analyze_errors(df_evaluated: pd.DataFrame) -> pd.DataFrame:
    """Analyze prediction errors and return wrong predictions"""
    
    # Filter rows where accuracy is 0 (wrong predictions)
    wrong_predictions = df_evaluated[df_evaluated['accuracy_score'] == 0].copy()
    
    if len(wrong_predictions) == 0:
        print("✅ All predictions are correct!")
        return pd.DataFrame()  # return empty DataFrame
    
    print(f"Found {len(wrong_predictions)} incorrect predictions "
          f"({len(wrong_predictions)/len(df_evaluated)*100:.1f}%)")
    
    # Display wrong predictions with relevant metrics
    display(wrong_predictions[['السياق بأكمله', 'المتوقع', 'predicted_token', 
                              'accuracy_score', 'synonym_match_score', 
                              'root_match_score', 'cosine_similarity_score']])
    
    # Categorize errors
    print("\nError Categories:")
    print("-" * 40)
    
    synonym_correct = wrong_predictions[wrong_predictions['synonym_match_score'] == 1]
    print(f"1. Synonym matches: {len(synonym_correct)} "
          f"({len(synonym_correct)/len(wrong_predictions)*100:.1f}%)")
    
    root_correct = wrong_predictions[wrong_predictions['root_match_score'] == 1]
    print(f"2. Root matches: {len(root_correct)} "
          f"({len(root_correct)/len(wrong_predictions)*100:.1f}%)")
    
    semantic_similar = wrong_predictions[wrong_predictions['cosine_similarity_score'] > 0.5]
    print(f"3. High semantic similarity (>0.5): {len(semantic_similar)} "
          f"({len(semantic_similar)/len(wrong_predictions)*100:.1f}%)")
    
    complete_misses = wrong_predictions[
        (wrong_predictions['synonym_match_score'] == 0) &
        (wrong_predictions['root_match_score'] == 0) &
        (wrong_predictions['cosine_similarity_score'] < 0.3)
    ]
    print(f"4. Complete misses: {len(complete_misses)} "
          f"({len(complete_misses)/len(wrong_predictions)*100:.1f}%)")
    
    if len(complete_misses) > 0:
        print("\nExamples of complete misses:")
        for idx, row in complete_misses.head(3).iterrows():
            print(f"\nContext: {row['السياق بأكمله']}")
            print(f"Expected: {row['المتوقع']}")
            print(f"Predicted: {row['predicted_token']}")
            print(f"Cosine Similarity: {row['cosine_similarity_score']:.3f}")
    
    return wrong_predictions

# Run error analysis and get only wrong predictions
wrong_predictions_df = analyze_errors(df_evaluated)
```

```python
# Cell 16: Export Results
def export_results(df_evaluated: pd.DataFrame, summary_df: pd.DataFrame, 
                   category_stats: pd.DataFrame = None, category1_stats: pd.DataFrame = None):
    """Export evaluation results to files"""
    
    timestamp = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
    
    # Export detailed results
    detailed_file = f"lm_evaluation_detailed_{timestamp}.csv"
    df_evaluated.to_csv(detailed_file, index=False, encoding='utf-8-sig')
    print(f"✓ Detailed results saved to: {detailed_file}")
    
    # Export summary
    summary_file = f"lm_evaluation_summary_{timestamp}.csv"
    summary_df.to_csv(summary_file, encoding='utf-8-sig')
    print(f"✓ Summary saved to: {summary_file}")
    
    # Export category stats if available
    category_file = None
    if category_stats is not None:
        category_file = f"lm_evaluation_by_category_{timestamp}.csv"
        category_stats.to_csv(category_file, encoding='utf-8-sig')
        print(f"✓ Category-wise performance saved to: {category_file}")
        
    if category1_stats is not None:
        category1_file = f"lm_evaluation_by_category1_{timestamp}.csv"
        category1_stats.to_csv(category1_file, encoding='utf-8-sig')
        print(f"✓ Category1-wise performance saved to: {category1_file}")
    else:
        category1_file = None
    
    # Export as Excel with multiple sheets
    excel_file = f"lm_evaluation_report_{timestamp}.xlsx"
    with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
        df_evaluated.to_excel(writer, sheet_name='Detailed_Results', index=False)
        summary_df.to_excel(writer, sheet_name='Summary', index=True)
        
        if category_stats is not None:
            category_stats.to_excel(writer, sheet_name='By_Category', index=True)
        
        if category1_stats is not None:
            category1_stats.to_excel(writer, sheet_name='By_Category1', index=True)
        
        # Add error analysis sheet
        wrong_predictions = df_evaluated[df_evaluated['accuracy_score'] == 0]
        if len(wrong_predictions) > 0:
            wrong_predictions.to_excel(writer, sheet_name='Error_Analysis', index=False)
    
    print(f"✓ Excel report saved to: {excel_file}")
    
    return detailed_file, summary_file, category_file, category1_file, excel_file

# Example usage
export_results(df_evaluated, summary_df, category_stats, category1_stats)

```

```python
# Cell 17: Model Comparison (if you have multiple models)
def compare_models(model_results: Dict[str, pd.DataFrame]):
    """Compare multiple models"""
    
    if len(model_results) < 2:
        print("Need at least 2 models for comparison")
        return
    
    comparison_data = []
    
    for model_name, df_model in model_results.items():
        # Calculate metrics for each model
        metrics = ['accuracy_score', 'synonym_match_score', 'root_match_score', 'cosine_similarity_score']
        model_metrics = {}
        
        for metric in metrics:
            if metric in df_model.columns:
                mean_score = df_model[metric].mean() * 100
                success_rate = (df_model[metric] > 0).mean() * 100
                model_metrics[f'{metric}_mean'] = mean_score
                model_metrics[f'{metric}_success'] = success_rate
        
        model_metrics['Model'] = model_name
        model_metrics['Num_Samples'] = len(df_model)
        comparison_data.append(model_metrics)
    
    comparison_df = pd.DataFrame(comparison_data)
    
    print("Model Comparison:")
    print("=" * 80)
    display(comparison_df)
    
    # Create comparison visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Model Comparison', fontsize=16, fontweight='bold')
    
    metrics_to_plot = ['accuracy_score', 'synonym_match_score', 'root_match_score', 'cosine_similarity_score']
    plot_titles = ['Accuracy', 'Synonym Match', 'Root Match', 'Cosine Similarity']
    
    for idx, (metric, title) in enumerate(zip(metrics_to_plot, plot_titles)):
        row = idx // 2
        col = idx % 2
        ax = axes[row, col]
        
        if f'{metric}_mean' in comparison_df.columns:
            # Create bar chart
            models = comparison_df['Model']
            scores = comparison_df[f'{metric}_mean']
            
            bars = ax.bar(models, scores, color=sns.color_palette("husl", len(models)))
            ax.set_title(f'{title} Comparison', fontsize=14, fontweight='bold')
            ax.set_ylabel('Mean Score (%)')
            ax.set_ylim(0, max(scores) * 1.2)
            
            # Add value labels
            for bar, score in zip(bars, scores):
                height = bar.get_height()
                ax.text(bar.get_x() + bar.get_width()/2., height + 1,
                       f'{score:.1f}%', ha='center', va='bottom', fontweight='bold')
            
            ax.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    return comparison_df

# Example of how to use model comparison
print("\nModel Comparison Example:")
print("-" * 40)
print("""
# To compare multiple models, use:
model_results = {
    'Fanar-LLM': df_evaluated,  # Your current results
    'Model-2': pd.read_csv('model2_predictions.csv', delimiter=';'),
    'Model-3': pd.read_csv('model3_predictions.csv', delimiter=';'),
}

comparison_df = compare_models(model_results)
""")
```
